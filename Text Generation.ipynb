```python
!pip install transformers
!pip install torch
!pip install sentencepiece

from transformers import pipeline

# 1. Extractive Summarization
from spacy.lang.en import English
from spacy.lang.en.stop_words import STOP_WORDS

def extractive_summarize(text, num_sentences=3):
    nlp = English()
    doc = nlp(text)
    sentences = list(doc.sents)
    
    # You'd implement a scoring mechanism here (e.g., TF-IDF) to select important sentences
    # For simplicity, let's just return the first few sentences
    return " ".join([str(sent) for sent in sentences[:num_sentences]])

text = """
The quick brown fox jumps over the lazy dog. This is another sentence. And yet another.
"""
print("Extractive Summary:", extractive_summarize(text))


```

```python
# 2. Abstractive Summarization
def abstractive_summarize(text):
    summarizer = pipeline("summarization") # Default model is often good
    summary = summarizer(text, max_length=100, min_length=30, do_sample=False) # You can tune these
    return summary[0]['summary_text']

text_to_summarize = """
The Eiffel Tower is a wrought-iron lattice tower located on the Champ de Mars in Paris, France. 
It is named after the engineer Gustave Eiffel, whose company designed and built the tower.
"""
print("Abstractive Summary:", abstractive_summarize(text_to_summarize))
```
